```python
import pandas as pd
df=pd.read_csv("mobile_sales.csv")
print("DataFrame loaded from CSV file:")
print(df.to_string())
```

**Q1. Display the summary of the Mobile Sales.**


```python
print("\nSummary of the Mobile Sales:")
print(df.info())
```

**Q2.Display the brands having prices over 1000 whose units sold is more than 40.**


```python
print("\nFiltering rows where the prices > 1000 and the units sold > 40:")
print(df[(df['Price'] > 1000) & (df['UnitsSold'] > 40)].to_string())
```

**Q3.Find the brands whose total units sold is more than 60.**

```python
print("\nFiltering brands whose toatl units sold is more than 60")
print(df[df['UnitsSold']>60])
```

**Q4.Find the mobile phones having prices between 600 and 1000.**

```python
print("\nFiltering rows where the prices of the mobile phones is in between 600 and 1000:")
print(df[(df['Price'] > 600) & (df['Price'] <= 1000)].to_string())
```

**Q5.Find the mobile phones which are paid in online mode.**

```python
print("\nFiltering rows where the mobile phones are bought through online mode:")
print(df['PaymentMethod']=='Online')
```

**Q6.Find out the buyers of mid age (i.e age between 30 and 50).**

```python
print("\nFiltering rows where the buyers are of mid age:")
print(df[(df['CustomerAge'] >= 30) & (df['CustomerAge'] <= 50)].to_string())
```

**Q7.Find out the brands whose revenue is more than 10000.**

```python
print("\nFiltering brands whose revenue is more than 10000")
print(df[df['TotalRevenue']>10000])
```

**Q8.Find out the brands bought by female mid agers.**

```python
print("\nFiltering rows having mid age female buyers:")
print(df[(df['CustomerAge'] >= 30) & (df['CustomerAge'] <= 50) & (df['CustomerGender'] == 'Female')].to_string())
```

**Q9.Find out the males who paid through debit card and credit card to buy mobile phones.**

```python
print("\nFiltering rows having males who used debit and credit card:")
print(df[(df['PaymentMethod'] == 'Credit Card') & (df['PaymentMethod'] == 'Debit Card') & (df['CustomerGender'] == 'Male')].to_string())
```

**Q10.Generate a descriptive statistics of the mobile sales.**

```python
print("\nDescriptive statistics of the mobile sales in 2024:")
print(df.describe())
```
