```python
import pandas as pd
df=pd.read_csv("combined_data_laliga.csv")
print("DataFrame loaded from CSV file:")
print(df.to_string())
```

**Q1.Display the summary of the La Liga 2014 Dataset**

```python
print("\nSummary of the La Liga Tournament:")
print(df.info())
```

**Q2.Select the rows of Away Team Corners and Away Team Throw Ins**

```python
print("\nSelecting columns 'Away Team Corners' and 'Away Team Throw Ins':")
print(df[['Away Team Corners' , 'Away Team Throw Ins']].to_string())
```

**Q3.Find the teams who have home team goals scored > 2 and away team goals scored <= 1.**

```python
print("\nFiltering rows where home team goals scored > 2 & away team goals scored <= 1:")
print(df[(df['Home Team Goals Scored'] > 2) & (df['Away Team Goals Scored'] <= 1)].to_string())
```

**Q4. Transform the home team fouls into two-third of it.**

```python
print("\nApplying a lambda function to make the home team fouls into two-third of it:")
print(df['Home Team Fouls'].apply(lambda x:x*(2/3)).to_string())
```

**Q5.Replace the value of Home Team Blocked Shots from 1 to 2.**

```python
print("\nReplace the value from 1 to 2 in 'Home Team Blocked Shots':")
print(df.replace(1,2).to_string)
```

**Q6.Plot Match Excitement over Home Team Rating**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
plt.plot(df.index,df['Match Excitement'],label='Match Excitement')
plt.xlabel('Home Team Rating')
plt.ylabel('Match Excitement')
plt.title('Match Excitement over Home Team Rating')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
